```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your dataset
data = pd.read_csv('Ready.csv') 

data.head()
```

```python
data.describe()
data.info()
```

```python
#Univariate Analysis 
```

```python
import matplotlib.pyplot as plt

# Histograms for numerical variables
numerical_columns = ['age', 'capital-gain', 'capital-loss', 'hours-per-week']
for column in numerical_columns:
    plt.figure(figsize=(8, 4))
    plt.hist(data[column], bins=20)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.show()
```

```python
# Box plots for numerical variables
plt.figure(figsize=(10, 6))
sns.boxplot(data=data['age'], orient="h")
plt.title("Box Plots for age")
plt.show()
```

```python
# Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data[['capital-loss', 'capital-gain']], fill=True, cbar=True)
plt.xlabel('Capital Loss')
plt.ylabel('Capital Gain')
plt.title('2D Kernel Density Plot:Capital Loss vs. Capital Gain')
plt.show()
```

```python
# Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data[['age', 'hours-per-week']], fill=True, cbar=True)
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.title('2D Kernel Density Plot: Age vs. Hours per Week')
plt.show()
```

```python
#Bivariate Analysis 
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='capital-gain', y='age', data=data)
plt.title('Line Chart: Age vs. Capital Gain')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='capital-loss', y='age', data=data)
plt.title('Line Chart: Age vs. Capital Loss')
plt.show()
```

```python
import seaborn as sns

# Pairwise correlation heatmap
corr_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Scatter plots for numeric variables
for col1 in numerical_columns:
    for col2 in numerical_columns:
        if col1 != col2:
            plt.figure(figsize=(6, 4))
            plt.scatter(data[col1], data[col2], alpha=0.5)
            plt.xlabel(col1)
            plt.ylabel(col2)
            plt.title(f'Scatter Plot of {col1} vs {col2}')
            plt.show()
```

```python
#violin plot
#A combination of a box plot and a kernel density plot. It displays the distribution of data across different categories.
sns.violinplot(x='education_Masters', y='age', data=data)
plt.title('Violin Plot of Age by Education')
plt.show()
```

```python
# Example violin plot to compare salary distributions by gender
plt.figure(figsize=(10, 6))
sns.violinplot(x='gender_Male', y='capital-gain', data=data, inner='quart')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.title('Violin Plot: Salary by Gender')
```

```python
# Pair Plot: 
#Scatter plots and histograms for all pairs of numerical variables.

sns.pairplot(data[['age', 'capital-gain', 'hours-per-week']])
plt.show()
```

```python
# pairplot with a subset of columns
subset = data[['capital-gain', 'capital-loss', 'gender_Female', 'gender_Male']]
sns.pairplot(subset)
```

```python
sns.countplot(x='race_White', data=data)
plt.title('Count of White vs. Non-White Individuals')
plt.show()
```

```python
from pandas.plotting import scatter_matrix

# Example scatterplot matrix with a subset of columns
subset = data[['age', 'capital-gain', 'capital-loss', 'hours-per-week']]
scatter_matrix(subset, alpha=0.2, figsize=(10, 6), diagonal='hist')
```

```python
# Select categorical columns to plot
categorical_cols = ['education_Bachelors', 'education_Masters', 'education_Doctorate']

# Create a box plot
plt.figure(figsize=(12, 6))
for col in categorical_cols:
    sns.boxplot(x=col, y='age', data=data)
plt.xticks(rotation=45)
plt.title('Box Plot for Categorical Variables')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='education_Bachelors', data=data, palette='Set1')
plt.title('Distribution of Education Level (Bachelors)')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Subset the relevant columns
subset = data[[
    'education_10th', 'education_11th', 'education_12th', 'education_1st-4th',
    'education_5th-6th', 'education_7th-8th', 'education_9th', 'education_Assoc-acdm',
    'education_Assoc-voc', 'education_Bachelors', 'education_Doctorate', 'education_HS-grad',
    'education_Masters', 'education_Preschool', 'education_Prof-school', 'education_Some-college',
    'occupation_Adm-clerical', 'occupation_Armed-Forces', 'occupation_Craft-repair',
    'occupation_Exec-managerial', 'occupation_Farming-fishing', 'occupation_Handlers-cleaners',
    'occupation_Machine-op-inspct', 'occupation_Other-service', 'occupation_Priv-house-serv',
    'occupation_Prof-specialty', 'occupation_Protective-serv', 'occupation_Sales',
    'occupation_Tech-support', 'occupation_Transport-moving'
]]

# Set the figure size
plt.figure(figsize=(12, 8))

# Create the heatmap
sns.heatmap(subset.corr(), cmap='coolwarm', annot=False, fmt=".1f")

# Set the title
plt.title('Heatmap of Education Level vs. Occupation')

# Show the plot
plt.show()
```

```python
# Create the heatmap with diagonal values masked
mask = np.triu(np.ones_like(subset.corr(), dtype=bool))
plt.figure(figsize=(12, 8))
sns.heatmap(subset.corr(), cmap='coolwarm', annot=False, fmt=".1f", mask=mask)
plt.title('Heatmap of Education Level vs. Occupation')
plt.show()
```
